```python
game_list = [0, 1, 2]
#  Display the current state of the game list.
def display_game(game_list):
    print("Here is the current list!")
    print(game_list)
    
    
#Ask the user to choose a position in the list (0, 1, or 2).
def position_choice():
    choice = 'Wrong'
    while choice not in ['0', '1', '2']:
        choice = input("Enter your choice (0-2): ")
        if choice not in ['0', '1', '2']:
            print("Sorry, but you did not choose a valid position.")
    return int(choice)


# Ask the user for a value to place at the selected position in the game list.
def replacement_choice(game_list, position):
    user_placement = input("Enter a value to place at the selected position: ")
    game_list[position] = user_placement
    return game_list


#   Ask the user if they want to keep playing (YES or NO).Repeats the prompt until a valid response is given.
def gameon_choice():
    choice = 'Wrong'
    while choice not in ['YES', 'NO']:
        choice = input("Keep playing? (YES/NO): ")
        if choice not in ['YES', 'NO']:
            print("Sorry, I don't understand that. Please choose between YES or NO.")
    return choice == 'YES'

game_on = True
game_list = [0, 1, 2]

while game_on:
    display_game(game_list)     # Display the current game list
    position = position_choice()
    game_list = replacement_choice(game_list, position)
    display_game(game_list)     # Display the updated game list
    game_on = gameon_choice()
```
